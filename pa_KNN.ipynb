```python
#LOAD LIBRARIES
import numpy as np
import seaborn as sb
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#LOAD DATASET
data = sb.load_dataset('iris')
```

```python
#SEE THE STRUCTURE OF DATASET
data.head()
```

```python
#GENERATE RANDOM INDICES FOR 90% OF ROWS
np.random.seed(40)
ran = np.random.choice(data.index, size = int(0.9 * len(data)), replace = False)
ran
```

```python
#EDA
print(data.info())
print(data.describe())
```

```python
#NORMALIZATION USING MINMAXSCALER
s = MinMaxScaler()
data_norm = pd.DataFrame(s.fit_transform(data.iloc[: , 0:4]), columns = data.columns[0:4])
```

```python
#SUMMARIES
print("\n Original Data Summary:")
print(data.describe())
print("\n\n\n")
print("\n Normalized Data Summary:")
print(data_norm.describe())
```

```python
#EXTRACT TRAINING AND TESTING SETS
data_train = data_norm.iloc[ran, :]
data_test = data_norm.drop(ran)
data_target_cat = data.iloc[ran, 4]
data_test_cat = data.drop(ran)['species']
```

```python
#RUN KNN
knn = KNeighborsClassifier(n_neighbors = 11)
knn.fit(data_train, data_target_cat)
pr = knn.predict(data_test)
```

```python
#CONFUSION MATRIX
mat = confusion_matrix(data_test_cat, pr)
print("\n CONFUSION MATRIX\n")
mat
```

```python
#CALCULATE ACCURACY
acc = accuracy_score(data_test_cat, pr) * 100
print("\n ACCURACY : {:.2f}%".format(acc))
```
